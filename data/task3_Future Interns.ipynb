```python
# BINNING VALUE
df_copy=df.copy()
df_copy= df_copy.drop('Student ID', axis=1)
bins = [0, 3, 5, 7, 10]  # Define bin edges
labels = ['1','2','3','4'] # Define bin labels
for x in df_copy.columns:
  df_copy[x]=pd.cut(df_copy[x],bins=bins, labels=labels,right=True)

#df_copy.isna()
df_copy
```
